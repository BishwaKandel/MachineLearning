```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.001
    
    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_pred)])
        md = -(2/n)*np.sum(x*(y-y_pred))
        bd = -(2/n)*np.sum(y-y_pred)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {} , b {} , cost {} , iteration {} ".format(m_curr,b_curr,cost,i))
    
```

```python
gradient_descent(x,y)
```

```python
def grad_desc(math,cs):
    m_curr = b_curr = 0
    learning_rate= 0.0002
    iterations = 1000
    n = len(math)
    cost_prev = 0 
    for i in range(iterations):
        cs_pred = m_curr*math + b_curr
        cost = (1/n) * np.sum((cs - cs_pred)**2)
        bd = -(2/n) * np.sum( cs - cs_pred )
        md = -(2/n) * np.sum ( math * (cs - cs_pred))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if np.isclose(cost_prev , cost ,rtol =  1e-20):
            break 
        cost_prev = cost 
        print("m {} , b {} , cost {} , iteration {} ".format(m_curr,b_curr,cost,i))
    
```
